```python
import os
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
!pip install pymorphy2
!pip install -U pymorphy2-dicts-ru
import pymorphy2 as pm
morph = pm.MorphAnalyzer()

nltk.download('punkt')
```

```python
with open('/content/stopwords.txt', 'r', encoding='utf-8') as f:
    stop_words = f.read().splitlines()

stop_words.extend(stopwords.words('russian'))

stw = list(stop_words)
```

```python
import zipfile
from zipfile import ZipFile

with zipfile.ZipFile('/content/science.zip', 'r') as z:
  z.extractall('science')

print('Done')
```

```python
corpus_files = os.listdir('science/science')
```

```python
for ind, txt in enumerate(corpus_files):
    with open('science/science//' + txt, 'r', encoding='utf-8') as f:
        text = nltk.word_tokenize(f.read().lower())
        text = [token for token in text if len(token) > 2]
        text = [morph.parse(token)[0].normal_form for token in text if token not in stop_words]

    with open('science_new.txt', 'a', encoding='utf-8') as g:
        g.write('\n'.join(text))
```

```python
f = open('/content/science_new.txt')
corpus = f.readlines()
```

```python
corpus
```



---



```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(corpus, test_size=0.2)
```



---



```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(min_df=10)

train_corpus_bow = vect.fit_transform(train)
test_corpus_bow = vect.fit_transform(test)

print(train_corpus_bow.shape)
print(test_corpus_bow.shape)
```



---



```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(min_df=10)

train_corpus_tfidf = vectorizer.fit_transform(train)
test_corpus_tfidf = vectorizer.fit_transform(test)

print(train_corpus_tfidf.shape)
print(test_corpus_tfidf.shape)
```
